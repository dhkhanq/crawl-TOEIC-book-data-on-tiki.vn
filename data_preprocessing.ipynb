```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', 100)
```

```python
df = pd.read_csv('./data/data.csv')
```

```python
data = df.copy()
data
```

```python
data.info()
```

```python
# Lọc ra các sách về TOEIC
data[data['Tên'].str.contains('TOEIC|Toeic|toeic', regex=True)]
```

```python
# Xoá các hàng không phải sách toeic
data.drop(data[~data['Tên'].str.contains('TOEIC|Toeic', regex=True)].index , inplace=True)
```

```python
pd.set_option('display.max_rows', 270)
data
```

```python
data.info()
```

```python
# Xử lý cột 'Giá'
# Thay thế dấu chấm và kí tự ₫ ở các phần tử của cột 'Giá' bằng rỗng
data['Giá'] = data['Giá'].str.replace('[.₫]', '', regex=True)
# Ép kiểu cột 'Giá' từ object thành int
data['Giá'] = data['Giá'].astype(str).astype(int)
```

```python
# Đổi tên cột 'Giảm giá' thành 'Giảm giá (%)'
data.rename(columns = {'Giảm giá':'Giảm giá (%)'}, inplace = True) 
# Thay thế dấu '-' và kí tự '%' ở các phần tử của cột 'Giảm giá (%)' bằng rỗng
data['Giảm giá (%)'] = data['Giảm giá (%)'].str.replace('[-%]', '', regex=True)
# Gán các giá trị rỗng ở cột 'Giảm giá (%)' bằng 0 (Không giảm giá)
data.loc[data['Giảm giá (%)'].isna(), 'Giảm giá (%)'] = 0
# Ép kiểu cột 'Giảm giá (%)' từ object thành int
data['Giảm giá (%)'] = data['Giảm giá (%)'].astype(str).astype(int)
```

```python
data
```

```python
# Thêm cột 'Giá (Chưa giảm)' vào dataframe
# Giá chưa giảm = Giá đã giảm / 1 - % giảm giá
data['Giá (Chưa giảm)'] = round((data['Giá'] / (1 - (data['Giảm giá (%)']/100))))
```

```python
data[['Giá (Chưa giảm)', 'Giá', 'Giảm giá (%)']]
```

```python
# Xóa chuỗi 'Đã bán ' cho các phần tử của cột
data['Đã bán'] = data['Đã bán'].str.replace("Đã bán ", "")
# Thay thế các giá trị NaN ở cột 'Đã bán' bằng 0
data.loc[data['Đã bán'].isna(), 'Đã bán'] = 0
```

```python
# Lọc ra các sách Hackers TOEIC
data[data['Tên'].str.contains('Hackers|hackers', regex=True)]
```

```python
# Thay thế các giá trị NaN bằng 'DAVID CHO'
data.loc[data['Tên'].str.contains('Hackers|hackers', regex=True) & data['Tác giả'].isna(), 'Tác giả'] = 'DAVID CHO'
```

```python
# Lọc ra các sách YBM
data[data['Tên'].str.contains('YBM', regex=True)]
```

```python
# Thay thế các giá trị NaN bằng 'YBM TOEIC R&D'
data.loc[data['Tên'].str.contains('YBM', regex=True) & data['Tác giả'].isna(), 'Tác giả'] = 'YBM TOEIC R&D'
```

```python
# Thay thế các giá trị NaN ở cột 'Tác giả' và 'Công ty phát hành' bằng 'Unknown'
data.loc[data['Tác giả'].isna(), 'Tác giả'] = "Unknown"
data.loc[data['Công ty phát hành'].isna(), 'Công ty phát hành'] = "Unknown"
```

```python
# Thay thế các giá trị NaN ở cột 'Loại bìa' bằng 'Bìa mềm'
data.loc[data['Loại bìa'].isna(), 'Loại bìa'] = "Bìa mềm"
```
