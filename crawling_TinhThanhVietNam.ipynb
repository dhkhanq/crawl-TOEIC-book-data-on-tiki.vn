```python
import numpy as np
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
URL = "https://vi.wikipedia.org/wiki/T%E1%BB%89nh_th%C3%A0nh_Vi%E1%BB%87t_Nam"
response = requests.get(URL)
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup
```

```python
tables = soup.find_all('table', "wikitable")
```

```python
table = tables[0].find('tbody')
```

```python
rows = table.find_all('tr')
```

```python
rows[0]
```

```python
column_names = []
for i in rows[0].find_all('th'):
    print(i.text.strip())
    column_names.append(i.text.strip())
```

```python
column_names
```

```python
del rows[0]
```

```python
data = {column_name : [] for column_name in column_names}
```

```python
data
```

```python
for i in rows:
    for index, j in enumerate(i.find_all('td')):
        value = j.text.strip()
        data[column_names[index]].append(value)
        
        
```

```python
data.keys()
```

```python
data
```

```python
df = pd.DataFrame(data)
df.to_csv('Tinh_thanh_Viet_Nam.csv', index=False)
```
