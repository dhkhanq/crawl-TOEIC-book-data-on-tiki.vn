## Xử lý cột Author bị rỗng và bị lỗi

```python
# Kiểm tra các tên tác giả hiện có
df['Author'][df['Author'].notnull()]
```

```python
# Tên tác giả bị lỗi gồm có '.' 'Park Hye Young, ', 'Paul Edmunds, '
# Xử lý '.' bằng cách thay thế bằng NaN
# Xử lý 'Park Hye Young, ' bằng cách bỏ đi phần đuôi ', '
df_ = df.copy()

for i in range(len(df_['Author'])):
    if df_['Author'][i] == '.':
        a = df_['Author'][i]
        df_['Author'][i] = np.nan
        print("\"{}\" -----> \"{}\"".format(a, df_['Author'][i]))
    if str(df_['Author'][i]).endswith('\xa0'):
        b = df_['Author'][i]
        df_['Author'][i] = df_['Author'][i].strip().rstrip(',')
        print("\"{}\" -----> \"{}\"".format(b, df_['Author'][i]))
```

```python
# Kiểm tra các hàng bị thiếu tên tác giả
df_[df_['Author'].isnull()]
```

```python
# list chứa index của các giá trị trống
id_lst = [2, 3, 4, 9, 10, 12, 16, 21, 27, 36, 38, 39]

# list chứa tên tác giả thay thế cho các giá trị trống
author_lst = ["David Cho", "Nhiều tác giả", "Nhiều tác giả", "Lee Ki Taek", "Nhiều tác giả", "Lee Ki Taek", "YBM TOEIC R&D", "Nhiều tác giả", "Nhiều tác giả","First News","Nhiều tác giả","Nhiều tác giả"]

for a, b in zip(id_lst, author_lst):
        df_['Author'][a] = b
```

# Xử lý cột List Price và Discount Rate

```python
# replace null value of List Price to value from current Price
# Vì cột List Price là giá sản phẩm gốc, nên nếu sách nào không giảm giá thì giá trị của nó bằng giá bán hiện tại (Cột Current Price)
df_["List Price"].fillna(df_["Current Price"], inplace=True)
# replace null value of Discount Rate to '0%'
# Những ô bị trống là những sản phẩm không được giảm giá, nên discount rate = 0%
df_["Discount Rate"].fillna("-0%", inplace=True)
```
